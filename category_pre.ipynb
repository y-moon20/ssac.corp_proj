### 필요한 컬럼만 추출

```python
dfc = df[['content_id','click_label','name', 'keyword', 'category_id_1', 'category_id_2', 'category_id_3']]
```

```python
dfc
```

```python
click = dfc.groupby('category_id_1').sum()['click_label']
```

```python
click
```

### 카테고리별로 데이터 분리

```python
for c in cname:
    globals()['dfc{}'.format(c)] = dfc[dfc['category_id_1'] == c]
```

### 카테고리별 중복데이터 정리

```python
unique = []
rowcnt = []
for c in cname:
    unique.append(globals()['dfc{}'.format(c)]['content_id'].nunique())
    rowcnt.append(len(globals()['dfc{}'.format(c)]['content_id']))
```

```python
nunique = pd.DataFrame({'impcnt' : rowcnt,
                        'content' : unique,
                       'click' : click,
                       'ctr' : click/rowcnt*100})
```

```python
nunique
```
